```
import pandas as pd
df=pd.read_csv("/content/drive/My Drive/Flight_weather.csv")
df.dropna(how='any',inplace=True)
df
```

```
import sklearn.preprocessing as preprocessing
le=preprocessing.LabelEncoder()
df["airport"]=le.fit_transform(df["airport"].astype(str))
df["FlightDate"]=le.fit_transform(df["FlightDate"].astype(str))
df
```

```
df=df[df['ArrDel15'] == 1.0]
df
```

```
X=df.drop(["ArrDel15","ArrDelayMinutes","ArrTime"],axis=1)
y=df["ArrDelayMinutes"]
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```
X_train
```

```
from sklearn.linear_model import LinearRegression
lrmodel=LinearRegression()
```

```
lrmodel.fit(X_train,y_train)
```

```
lrmodel.coef_
```

```
lrmodel.intercept_
```

```
y_pred=lrmodel.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import pickle as pkl
```

```
print("rmse for linear regression is:",mean_squared_error(y_test,y_pred,squared=False))
print("MAE for linear regression is:",mean_absolute_error(y_test,y_pred))
print("R2_Score for linear regression is:",r2_score(y_test,y_pred))
```

```
from sklearn.ensemble import ExtraTreesRegressor
extra_tree_regressor = ExtraTreesRegressor()
```

```
extra_tree_regressor.fit(X_train,y_train)
```

```
y_pred_extratrees=extra_tree_regressor.predict(X_test)
```

```
print("rmse for extra trees regressor is:",mean_squared_error(y_test,y_pred_extratrees,squared=False))
print("MAE for extra trees regressor is:",mean_absolute_error(y_test,y_pred_extratrees))
print("R2_Score for extra trees regressor is:",r2_score(y_test,y_pred_extratrees))
```

```
from sklearn.ensemble import RandomForestRegressor
Rf_regressor = RandomForestRegressor()
```

```
Rf_regressor.fit(X_train,y_train)
pkl.dump(Rf_regressor,open("finalized_regreseeor.sav","wb"))
```

```
y_pred_RandomForest=Rf_regressor.predict(X_test)
```

```
print("rmse for random forest regressor is:",mean_squared_error(y_test,y_pred_RandomForest,squared=False))
print("MAE for random forest regressor is:",mean_absolute_error(y_test,y_pred_RandomForest))
print("R2_Score for random forest regressor is:",r2_score(y_test,y_pred_RandomForest))
```

```
from xgboost import XGBRegressor
Xgbregressor=XGBRegressor(objective="reg:linear")
```

```
Xgbregressor.fit(X_train,y_train)
```

```
y_pred_Xg=Xgbregressor.predict(X_test)
```

```
print("rmse for Xgboost regressor is:",mean_squared_error(y_test,y_pred_Xg,squared=False))
print("MAE for Xgboost regressor is:",mean_absolute_error(y_test,y_pred_Xg))
print("R2_Score for Xgboost regressor is:",r2_score(y_test,y_pred_Xg))
```
