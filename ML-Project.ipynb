```
import json
datasets=[None]*15
list_airports=["ATL","CLT","DEN","DFW","EWR","IAH","JFK","LAS","LAX","MCO","MIA","ORD","PHX","SEA","SFO"]
for j in range(15):
  for i in range(1,13):
    f=open("/content/drive/My Drive/Project/weather/"+list_airports[j]+"/2016-"+str(i)+".json","r")
    data_weather=json.load(f)
    datasets[j]=pd.concat([datasets[j],pd.DataFrame(data_weather['data']['weather'])])
    f.close()
    f=open("/content/drive/My Drive/Project/weather/"+list_airports[j]+"/2017-"+str(i)+".json","r")
    data_weather=json.load(f)
    datasets[j]=pd.concat([datasets[j],pd.DataFrame(data_weather['data']['weather'])])
  datasets[j]['airport']=list_airports[j]
df_weather=pd.concat(datasets)
df_weather
```

```
df2=df_weather.explode('hourly')
df2
```

```
from pandas.io.json import json_normalize
df3 = json_normalize(df2['hourly'])
```
