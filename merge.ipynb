```
import pandas as pd
```

```
df_flight=pd.read_csv("/content/drive/My Drive/flight_data.csv")
df_flight
```

```
df_weather=pd.read_csv("/content/drive/My Drive/weather_data.csv")
df_weather
```

```
Time=df_flight['DepTime'].tolist()
for i in range(len(Time)):
  hours=Time[i]//100
  mins=Time[i]%100
  if(mins > 30):
    hours+=1
  Time[i]=hours * 100 
df_flight['newroundedtime']=Time
df_flight
```

```
df_flight['newroundedtime']=df_flight['newroundedtime'].fillna(0.0).astype(int)
df_flight
```

```
df_flight
```

```
df_merged=pd.merge(df_flight,df_weather,left_on=['FlightDate','newroundedtime','Origin'],right_on=['date','time','airport']).drop(['date','newroundedtime','time','Origin'],axis=1)
df_merged
```
